```python
# ============================================
# Load the dataset - FIXED VERSION
# ============================================

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Try multiple possible file paths
possible_paths = [
    '../data/CVD Dataset.csv',      # Original name (with space)
    '../data/CVD_Dataset.csv',      # Without space
    'data/CVD Dataset.csv',         # If running from notebooks folder
    'data/CVD_Dataset.csv',         # Alternative
    '../CVD Dataset.csv',           # If data is in parent folder
    '../CVD_Dataset.csv',           # Alternative
]

file_loaded = False

for path in possible_paths:
    if os.path.exists(path):
        print(f" Found file at: {path}")
        df = pd.read_csv(path)
        file_loaded = True
        break

if not file_loaded:
    # If still not found, let's see what's actually there
    print(" File not found! Let's check your folder structure:")
    print(f"Current directory: {os.getcwd()}")
    
    # Check what's in the current directory
    print("\n Contents of current directory:")
    for item in os.listdir('.'):
        if os.path.isdir(item):
            print(f"    {item}/")
            # Show contents of subdirectories
            try:
                sub_items = os.listdir(item)
                for sub_item in sub_items:
                    print(f"       {sub_item}")
            except:
                pass
        else:
            print(f"    {item}")
    
    raise FileNotFoundError("Could not find CVD Dataset.csv in any expected location!")

print("="*50)
print("DATASET LOADED SUCCESSFULLY!")
print("="*50)
print(f"\n Shape: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
# ============================================
# Basic Dataset Information
# ============================================

print("\n" + "="*50)
print("BASIC DATASET INFORMATION")
print("="*50)

print(f"\n Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")

print("\n Column Names:")
for i, col in enumerate(df.columns, 1):
    print(f"   {i}. {col}")

print("\n Data Types:")
print(df.dtypes)

print("\n Statistical Summary (Numerical Columns):")
display(df.describe())

print("\n Dataset Info:")
df.info()
```

```python
# ============================================
# Missing Values Analysis
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("\n" + "="*50)
print("MISSING VALUES ANALYSIS")
print("="*50)

# Check missing values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_values,
    'Percentage': missing_percentage
})

# Filter columns with missing values
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

if len(missing_df) > 0:
    print("\nColumns with Missing Values:")
    display(missing_df)
    
    # Visualize missing values
    plt.figure(figsize=(10, 6))
    plt.bar(missing_df.index, missing_df['Missing Count'])
    plt.title('Missing Values by Column')
    plt.xlabel('Columns')
    plt.ylabel('Number of Missing Values')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('../images/missing_values.png')
    plt.show()
else:
    print("\nNo missing values found in the dataset!")
```

```python
# ============================================
# Duplicate Records Check
# ============================================

print("\n" + "="*50)
print("DUPLICATE RECORDS CHECK")
print("="*50)

duplicates = df.duplicated().sum()
print(f"\n Duplicate rows found: {duplicates}")

if duplicates > 0:
    print(f"   Removing {duplicates} duplicate rows...")
    df = df.drop_duplicates()
    print(f"   New dataset shape: {df.shape}")
else:
    print("    No duplicates found!")
```

```python
# ============================================
# Target Variable Analysis
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("\n" + "="*50)
print("TARGET VARIABLE ANALYSIS")
print("="*50)

# Assuming target column names (adjust based on your dataset)
target_col = 'CVD Risk Level' if 'CVD Risk Level' in df.columns else None

if target_col is None:
    # Try to find the target column
    for col in df.columns:
        if 'risk' in col.lower() or 'target' in col.lower() or 'level' in col.lower():
            target_col = col
            break

if target_col:
    print(f"\nTarget Variable: {target_col}")
    
    # Value counts
    print("\nDistribution:")
    target_dist = df[target_col].value_counts()
    print(target_dist)
    
    # Percentage distribution
    print("\nPercentage Distribution:")
    target_percent = df[target_col].value_counts(normalize=True) * 100
    print(target_percent)
    
    # Visualize target distribution
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Count plot
    ax1 = sns.countplot(data=df, x=target_col, ax=axes[0])
    axes[0].set_title(f'Distribution of {target_col}')
    axes[0].set_xlabel(target_col)
    axes[0].set_ylabel('Count')
    
    # Add count labels on bars
    for p in ax1.patches:
        ax1.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom')
    
    # Pie chart
    df[target_col].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%')
    axes[1].set_title(f'Percentage Distribution of {target_col}')
    axes[1].set_ylabel('')
    
    plt.tight_layout()
    plt.savefig('../images/target_distribution.png')
    plt.show()
    
    # Check for class imbalance
    min_class = target_percent.min()
    max_class = target_percent.max()
    imbalance_ratio = max_class / min_class
    
    print(f"\nClass Imbalance Ratio: {imbalance_ratio:.2f}")
    if imbalance_ratio > 2:
        print("   Dataset is imbalanced! Consider using SMOTE or class weights.")
    else:
        print("   Dataset is relatively balanced.")
else:
    print("Could not identify target column automatically.")
    print(f"Available columns: {df.columns.tolist()}")
```

```python
# ============================================
# Numerical Features Analysis
# ============================================

print("\n" + "="*50)
print("NUMERICAL FEATURES ANALYSIS")
print("="*50)

# Identify numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Remove target if it's numerical
if target_col in numerical_cols:
    numerical_cols.remove(target_col)

print(f"\n Numerical Columns ({len(numerical_cols)}):")
print(numerical_cols)

if len(numerical_cols) > 0:
    # Distribution plots
    n_cols = min(len(numerical_cols), 4)
    n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes]
    
    for i, col in enumerate(numerical_cols):
        if i < len(axes):
            ax = axes[i]
            sns.histplot(df[col], kde=True, ax=ax)
            ax.set_title(f'Distribution of {col}')
            ax.set_xlabel(col)
            ax.set_ylabel('Frequency')
            
            # Add statistics
            mean_val = df[col].mean()
            median_val = df[col].median()
            ax.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.1f}')
            ax.axvline(median_val, color='green', linestyle='--', label=f'Median: {median_val:.1f}')
            ax.legend()
    
    # Hide unused subplots
    for i in range(len(numerical_cols), len(axes)):
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.savefig('../images/numerical_distributions.png')
    plt.show()
    
    # Correlation heatmap
    if len(numerical_cols) > 1:
        plt.figure(figsize=(12, 10))
        correlation = df[numerical_cols].corr()
        mask = np.triu(np.ones_like(correlation, dtype=bool))
        sns.heatmap(correlation, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r', 
                   center=0, square=True, linewidths=0.5)
        plt.title('Correlation Matrix of Numerical Features')
        plt.tight_layout()
        plt.savefig('../images/correlation_matrix.png')
        plt.show()
        
        # Find highly correlated features
        high_corr = []
        for i in range(len(correlation.columns)):
            for j in range(i+1, len(correlation.columns)):
                if abs(correlation.iloc[i, j]) > 0.8:
                    high_corr.append((correlation.columns[i], correlation.columns[j], 
                                    correlation.iloc[i, j]))
        
        if high_corr:
            print("\n Highly Correlated Features (>0.8):")
            for feat1, feat2, corr in high_corr:
                print(f"   {feat1} ↔ {feat2}: {corr:.2f}")
            print("   Consider removing one to avoid multicollinearity.")
```

```python
# ============================================
# Categorical Features Analysis
# ============================================

print("\n" + "="*50)
print("CATEGORICAL FEATURES ANALYSIS")
print("="*50)

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

if target_col in categorical_cols:
    categorical_cols.remove(target_col)

print(f"\n Categorical Columns ({len(categorical_cols)}):")
print(categorical_cols)

if len(categorical_cols) > 0:
    n_cols = min(len(categorical_cols), 3)
    n_rows = (len(categorical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if n_rows > 1 else [axes]
    
    for i, col in enumerate(categorical_cols):
        if i < len(axes):
            ax = axes[i]
            value_counts = df[col].value_counts()
            bars = ax.bar(value_counts.index, value_counts.values)
            ax.set_title(f'Distribution of {col}')
            ax.set_xlabel(col)
            ax.set_ylabel('Count')
            ax.tick_params(axis='x', rotation=45)
            
            # Add value labels
            for bar in bars:
                height = bar.get_height()
                ax.annotate(f'{int(height)}',
                           xy=(bar.get_x() + bar.get_width() / 2, height),
                           xytext=(0, 3),
                           textcoords="offset points",
                           ha='center', va='bottom')
    
    # Hide unused subplots
    for i in range(len(categorical_cols), len(axes)):
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.savefig('../images/categorical_distributions.png')
    plt.show()
```

```python
# ============================================
# Quick Feature Importance using Random Forest
# ============================================

print("\n" + "="*50)
print("INITIAL FEATURE IMPORTANCE")
print("="*50)

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Prepare data for quick importance check
if target_col:
    # Create a copy for quick analysis
    df_quick = df.copy()
    
    # Encode categorical columns
    le = LabelEncoder()
    for col in categorical_cols:
        df_quick[col] = le.fit_transform(df_quick[col].astype(str))
    
    # Prepare X and y
    X = df_quick.drop(columns=[target_col])
    y = df_quick[target_col]
    
    # Encode target if needed
    if y.dtype == 'object':
        y = le.fit_transform(y)
    
    # Quick Random Forest
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X, y)
    
    # Get feature importance
    importance_df = pd.DataFrame({
        'Feature': X.columns,
        'Importance': rf.feature_importances_
    }).sort_values('Importance', ascending=False)
    
    print("\n Top 10 Most Important Features:")
    display(importance_df.head(10))
    
    # Visualize feature importance
    plt.figure(figsize=(10, 8))
    top_features = importance_df.head(10)
    plt.barh(top_features['Feature'], top_features['Importance'])
    plt.title('Top 10 Important Features (Initial Estimate)')
    plt.xlabel('Importance')
    plt.tight_layout()
    plt.savefig('../images/initial_feature_importance.png')
    plt.show()
else:
    print(" Target column not found. Skipping feature importance.")
```

```python
# ============================================
# Summary Statistics and Key Insights
# ============================================

print("\n" + "="*50)
print("SUMMARY STATISTICS")
print("="*50)

print(f"\n Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")
print(f" Numerical Features: {len(numerical_cols)}")
print(f" Categorical Features: {len(categorical_cols)}")
print(f" Target Variable: {target_col}")

if target_col:
    print(f"\n Target Distribution:")
    print(df[target_col].value_counts())

print(f"\n Missing Values:")
print(df.isnull().sum().sum())

print("\n Data Analysis Complete!")
print("\n Visualizations saved in '../images/' folder")
print(" Key Findings:")
print("   - Check the target distribution for class imbalance")
print("   - Review correlation matrix for highly correlated features")
print("   - Note important features from Random Forest analysis")
```
